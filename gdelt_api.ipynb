__Novetta asked if every article (source url) in the mentions table is an article in the events table. To find out, for every article we'll look at the corresponding globaleventid. Then we'll get all articles corresponding to that globaleventid in the events table, and we'll see if one of those source url's is the sourceurl we started with from the mentions table.__

note that eventtimedate in the mentions table corresponds to dateadded in the events table (according to the documentation)

__Note that dates are stored as ints__

__To answer the question, "are there articles (MentionIdentifiers) in the mentions table that don't have corresponding articles (SOURCEURL) in the events table, we'll do a right join.__

```python
merge_source = pd.merge(results, mentions, left_on='SOURCEURL', right_on = 'MentionIdentifier', how='right')
merge_source.head()
```

```python
merge_source[merge_source['GLOBALEVENTID_x'].isnull()].head()
```
